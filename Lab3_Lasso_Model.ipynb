# 2.) Come up with 6 search terms you think could nowcast earnings. (Different than the ones I used) Add in 3 terms that that you think will not Nowcast earnings. Pull in the gtrends data

```python
from pytrends.request import TrendReq
```

```python
# Create pytrends object
pytrends = TrendReq(hl='en-US', tz=360)

# Set up the keywords and the timeframe
keywords = ["China", "Steve Jobs", "Camera", "Tim Cook", "September", "iTunes", "Renaissance", "Restaurant", "FSU"]
start_date = '2004-01-01'
end_date = '2024-01-01'

# Create an empty DataFrame to store the results
df = pd.DataFrame()

# Iterate through keywords and fetch data
for keyword in keywords:
    pytrends.build_payload([keyword], cat=0, timeframe=f'{start_date} {end_date}', geo='', gprop='')
    interest_over_time_df = pytrends.interest_over_time()
    df[keyword] = interest_over_time_df[keyword]
```

```python
df = df.resample("Q").mean()
df
```

```python
temp = pd.concat([y, df],axis = 1).dropna()

y = temp[["BasicEPS"]].copy()
X = temp.iloc[:,1:].copy()

temp
```

# 3.) Normalize all the X data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

# 4.) Run a Lasso with lambda of .5. Plot a bar chart.

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha = 0.5)
```

```python
lasso_model = lasso.fit(X_scaled,y)
```

```python
coefficients = lasso_model.coef_
coefficients
```

```python
plt.figure(figsize=(12, 5))
plt.bar(range(len(coefficients)), coefficients)
plt.axhline(0, color="red", linestyle="--", linewidth=2)
plt.xticks(range(len(coefficients)), X.columns, rotation=45, ha="right")
plt.title('Lasso Coefficients')
plt.show()
```

# 5.) Do these coefficient magnitudes make sense?

At a lambda of 0.5, there is moderate regularization happening. The model has decided that none of these features have sufficient predictive power and pushes their coefficients to zero. These coefficients make sense given the lambda, and a lower lambda is probably best.
